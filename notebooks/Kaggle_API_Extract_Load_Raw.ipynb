```python
# Kaggle_API_Extract_Load_Raw.ipynb

# STEP 1: Install necessary packages
%pip install kaggle pandas sqlalchemy psycopg2-binary python-dotenv

# STEP 2: Import libraries
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

# STEP 3: Load environment variables
load_dotenv()

DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASS = os.getenv("DB_PASS")

# STEP 4: Kaggle API setup
# Assuming you've placed kaggle.json in ~/.kaggle already.
!kaggle datasets download -d christophertreasure/mlb-team-payrolls-2011-2024

import zipfile
with zipfile.ZipFile("mlb-team-payrolls-2011-2024.zip", "r") as zip_ref:
    zip_ref.extractall("mlb_payroll_data")

# STEP 5: Read the CSV data
df = pd.read_csv("/workspaces/moneyball/notebooks/data/mlb_payrolls.csv")
print("First few rows of the data:")
print(df.head())

# STEP 6: Clean / inspect data (adjust this as needed)
print("Columns and null values:")
print(df.info())
print(df.isnull().sum())

# STEP 7: Connect to PostgreSQL
connection_string = f"postgresql+psycopg2://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(connection_string)

# STEP 8: Write dataframe to SQL (sql_project schema)
df.to_sql("mlb_team_payrolls", engine, schema="sql_project", if_exists="replace", index=False)

print("Data uploaded successfully to sql_project.mlb_team_payrolls!")

# STEP 9: Verify the data
result = pd.read_sql("SELECT * FROM sql_project.mlb_team_payrolls LIMIT 10;", con=engine)
print("Sample rows from the database:")
print(result)
```
